# Web Scrapping on Fruit Box & Co. Website.

- We are going to scrape https://fruitboxco.com/collections/daily-fruits
- we will extract details about each motor bike.
- we will make a csv file using details like name, price, cc, mileage,weight link

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
project_url='https://fruitboxco.com/collections/daily-fruits'
```

```python
res= requests.get(project_url)
```

```python
len(res.text)
```

```python
res.status_code
```

```python
content= res.text
```

```python
content[:1000]
```

```python
doc= bs(content,"html.parser")
```

```python
fruit_url=  doc.find_all('h2',{'class':"ProductItem__Title Heading"})
fruit_url_tags='fruitboxco.com'+fruit_url[0].a['href']
fruit_url_tags
```

```python
fruit_name_tags=doc.find_all('h2')
fruit_name_tags[0].text.strip() #fruit_name_tags[0].a['href']
```

```python
discounted_price_tags= doc.find_all('span',{'class':"ProductItem__Price Price Price--highlight Text--subdued"})
discounted_price_tags[0].text.strip()
```

```python
original_price_tags=doc.find_all('span',{'class':'ProductItem__Price Price Price--compareAt Text--subdued'})
original_price_tags[0].text.strip()
```

```python
fruit_details=[]
for link,name,dp,op in zip(fruit_url,fruit_name_tags,discounted_price_tags,original_price_tags):
    fruit_details.append({'Fruit Name':name.text.strip(),'Original price':op.text.strip(),'Discounted price':dp.text.strip(),'Link':'fruitboxco.com'+link.a['href']})
```

```python
fruit_details
```

```python
df=pd.DataFrame(fruit_details)
df.head()
```

```python
df.to_csv('Fruit box data.csv')
```
